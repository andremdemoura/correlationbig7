```python
# Import libraries

import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Download historical adjusted closing prices from BIG7 stocks and Bitcoin (starting from 2020-01-01)

stocks = yf.download("AAPL MSFT AMZN GOOG META NVDA TSLA BTC-USD", start="2020-01-01",auto_adjust=False)['Adj Close']
```

```python
# Calculate daily returns and remove NaN values

returns = stocks.pct_change().dropna()
```

```python
# Compute the correlation matrix

correlation = returns.corr()
```

```python

# Plot the correlation matrix using a heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between BIG7")
plt.show()
```
